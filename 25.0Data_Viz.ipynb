## Libraries

## Bar Chart

```python
df_used_cars = pd.read_csv('bar_chart_data.csv')
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x= df_used_cars['Brand'], height= df_used_cars['Cars Listings'],
       color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title("Car Listings by Brand", fontsize = 16, fontweight = 'bold' )
plt.ylabel("Number of Listings", fontsize= 13)
plt.savefig("Used_Cars_Bar.png")
plt.show()
```

## Pie Chart

## Stacked Area Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
```

```python
df_fuel_engine_types
```

```python
colors = ["#011638","#7e2987", '#ef2026']
labels = ['Diesel', 'Petrol', 'Gas']
sns.set_style('white')
plt.figure(figsize =(12,6))
plt.stackplot(df_fuel_engine_types['Year'],
               df_fuel_engine_types['Diesel'],
               df_fuel_engine_types['Petrol'],
               df_fuel_engine_types['Gas'],
             colors = colors,
             edgecolor = 'none')
plt.xticks(df_fuel_engine_types['Year'],rotation = '45')
plt.legend(labels = labels, loc = 'upper left')
plt.ylabel('Number of cars', fontsize = 13)
plt.title('Popularity of Engine Fuel Types(1982-2016)', fontsize = '14', weight = 'bold')
sns.despine() #TOP = True,right = True
plt.show()
```

## Scatter Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("scatter_data.csv")
```

```python
df_real_estate
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(df_real_estate['Area (ft.)'], df_real_estate['Price'], 
           alpha= 0.6,
           c =df_real_estate['Building Type'],
           cmap = 'viridis')
plt.legend(*scatter.legend_elements(), loc = "upper left", title = "Building Type")
plt.title("Relationship between Area and Price of California Real Estate", fontsize =14, weight ="bold")
plt.xlabel("Area (sq. ft)", weight = "bold")
plt.ylabel("Price (000's pf $)")
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.scatterplot(df_real_estate['Price'], df_real_estate['Area (ft.)'],
               hue =df_real_estate['Building Type'],
               palette =['black', 'darkblue', 'purple', 'pink', 'white'],
               s = 100)
plt.legend(*scatter.legend_elements(), loc = "upper left", title = "Building Type")
plt.title("Relationship between Area and Price of California Real Estate", fontsize =14, weight ="bold")
plt.xlabel("Area (sq. ft)", weight = "bold")
plt.ylabel("Price (000's pf $)")
plt.show()
```

## Regression Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_expenditure = pd.read_csv("scatter_plot_ii.csv")
```

```python
df_ad_expenditure
```

```python
#plt.figure(figsize = (10,8))
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x ="Budget", y = "Sales", data = df_ad_expenditure,
           scatter_kws = {'color':'k'}, line_kws ={'color': 'red'})
plt.xlabel("Ad Expenditure in (000' $)")
plt.ylabel("Sales in (000's units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = 'bold')
plt.show()
```

```python
sns.lmplot(x ="Budget", y = "Sales", data = df_ad_expenditure,
           scatter_kws = {'color':'k'}, line_kws ={'color': 'red'},
          height = 10)
plt.xlabel("Ad Expenditure in (000' $)")
plt.ylabel("Sales in (000's units)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = 'bold')
plt.show()
```

## Bar and Line Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.ticker import PercentFormatter
```

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
```

```python
df_kdnuggets
```

```python
fig, ax = plt.subplots(figsize =(10,7))
ax.bar(df_kdnuggets['Year'],df_kdnuggets['Participants'],
      color = 'k')
ax.set_ylabel("Number of Participants", weight = 'bold')
ax.tick_params(axis ="y", width =2, labelsize = 'large')
ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kdnuggets['Year'], df_kdnuggets['Python Users'],
        color = '#b60000', marker = 'D')
ax1.set_ylabel("Python Users", color = '#b60000', weight = 'bold')
ax1.tick_params(axis = "y", width = 2, labelsize = 'large', colors = '#b60000')
ax.set_title("KD Nuggets Survey Python Users (2012 -2019)", fontsize = '14', weight = 'bold')
```
